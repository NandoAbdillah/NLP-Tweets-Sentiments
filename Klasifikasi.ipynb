# 1) Pembagian Data

```python
import random
def train_test_split(features, targets, test_size=0.2, random_state=None) :
    #Mengatur seed untuk reproducibility
    if random_state is not None :
        random.seed(random_state)

    #Mengacak data
    data = list(zip(features.items(), targets))
    random.shuffle(data)

    #Menghitung ukuran set pengujian
    test_size = int(len(data) *test_size)

    #Membagi data menjadi set
    test_data = data[:test_size]
    train_data = data[test_size:]

    #Mengurai fitur dan target
    train_features, train_targets = zip(*train_data)
    test_features , test_targets = zip(*test_data)

    return train_features, test_features, train_targets, test_targets

 
```

```python
# Import library yang dibutuhkan
import math

# Fungsi untuk menghitung jarak euclidean
def euclidean_distance(vec1, vec2):
    return math.sqrt(sum((vec1[key] - vec2.get(key, 0))**2 for key in vec1.keys()))

# Fungsi untuk mendapatkan k tetangga terdekat
def get_neighbors(train_features, test_feature, k):
    distances = [(i, euclidean_distance(train_feature[1], test_feature[1])) for i, train_feature in enumerate(train_features)]
    distances.sort(key=lambda x: x[1])
    neighbors = distances[:k]
    return [index for index, _ in neighbors]

# Fungsi untuk melakukan prediksi
def predict_classification(train_targets, neighbors):
    target_counts = Counter(train_targets[i] for i in neighbors)
    return target_counts.most_common(1)[0][0]

# Fungsi untuk melakukan klasifikasi dengan KNN
def knn(train_features, train_targets, test_features, k):
    predictions = []
    for test_feature in test_features:
        neighbors = get_neighbors(train_features, test_feature, k)
        prediction = predict_classification(train_targets, neighbors)
        predictions.append(prediction)
    return predictions

# Melakukan klasifikasi dengan KNN
k = 5
predictions = knn(train_features, train_targets, test_features, k)
```
